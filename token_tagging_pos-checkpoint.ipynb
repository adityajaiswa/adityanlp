```python
import nltk
text1='''His conclusion also contained the only serious part of his entire speech. Obama’s conclusion ended his speech perfectly. '''
for text in text1:
    sentences=nltk.sent_tokenize(text1)
    for sentence in sentences:
        words=nltk.word_tokenize(sentence)
        tagged=nltk.pos_tag(words)
        print(tagged)
```
